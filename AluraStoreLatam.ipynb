```python
## C칩digo del analisis para visualizar la evolucion de ingresos de toda la empresa en base a los a침os.

tienda_global=[tienda, tienda2, tienda3, tienda4]
df_tienda_global=pd.concat(tienda_global, ignore_index=True)
df_tienda_global['Fecha de Compra']=pd.to_datetime(df_tienda_global['Fecha de Compra'], dayfirst=True)
ingreso_anual=df_tienda_global.groupby(df_tienda_global['Fecha de Compra'].dt.year)['Precio'].sum().to_frame().reset_index()

## grafico

df_grafico = ingreso_anual.reset_index()

plt.figure(figsize=(12, 6))
plt.grid(True, linestyle='--', alpha=0.5)

plt.plot(df_grafico['Fecha de Compra'], df_grafico['Precio'],
         marker='o', linestyle='-', color='b', linewidth=2, markersize=8)

for i in range(len(df_grafico)):
    valor_real = df_grafico['Precio'].iloc[i]
    a침o = df_grafico['Fecha de Compra'].iloc[i]

    valor_millones = valor_real / 1000000
    etiqueta = f"${valor_millones:,.1f}M"

    plt.text(a침o, valor_real + (valor_real * 0.05), etiqueta,
             ha='center', va='bottom', fontsize=11, fontweight='bold', color='darkblue')

plt.ticklabel_format(style='plain', axis='y')
plt.title('Evoluci칩n de Ingresos (en Millones de $)', fontsize=14, pad=20)
plt.xlabel('A침o')
plt.ylabel('Ingresos Totales ($)')

plt.ylim(0, df_grafico['Precio'].max() * 1.3)

plt.show()
```
